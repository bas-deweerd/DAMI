```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import somoclu
#%matplotlib inline
```

```python
#help(np.random.rand)
np.random.seed(0)

c1 = np.random.rand(50, 3)/5
c2 = (0.6, 0.1, 0.05) + np.random.rand(50, 3)/5
c3 = (0.4, 0.1, 0.7) + np.random.rand(50, 3)/5

data = np.float32(np.concatenate((c1, c2, c3)))

#x1 = np.array([[1,2,3],
#               [4,5,6]])
#x2 = np.array([[10,11,12],
#               [13,14,15]])

#np.concatenate((x1,x2))

#print(np.concatenate((c1,c2,c3)))
colors = ["red"] * 50
colors.extend(["green"] * 50)
colors.extend(["blue"] * 50)

#help(plt.figure)

fig = plt.figure()
print(fig)
ax = Axes3D(fig)
print(ax)
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=colors)
print(ax)

#c1 = np.random.rand(50, 3)/5
#c2 = (0.6, 0.1, 0.05) + np.random.rand(50, 3)/5
#c3 = (0.4, 0.1, 0.7) + np.random.rand(50, 3)/5
#data = np.float32(np.concatenate((c1, c2, c3)))
#colors = ["red"] * 50
#colors.extend(["green"] * 50)
#colors.extend(["blue"] * 50)
#fig = plt.figure()
#ax = Axes3D(fig)
#ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=colors)
labels = range(150)
```

```python
number_of_rows, number_of_columns = 100, 160

#help(somoclu.Somoclu)

som = somoclu.Somoclu(number_of_rows, number_of_columns, data=data, kerneltype=0, verbose=2)
%time som.train()
```
