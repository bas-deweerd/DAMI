```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(1)
%matplotlib notebook

x = np.random.rand(15)
y = np.random.rand(15)

figure,axes = plt.subplots()
scatterPlot = plt.scatter(x,y, s=100)

annotation = plt.gca().annotate("", xy=(0,0), xytext=(20,20),textcoords="offset points",
                    bbox=dict(boxstyle="round", fc="w"),
                    arrowprops=dict(arrowstyle="->"))
annotation.set_visible(False)

def update_annotation(ind):
    position = scatterPlot.get_offsets()[ind["ind"][0]]
    annotation.xy = position
    annotation.set_text("Hello world")
    annotation.get_bbox_patch().set_alpha(0.4)    
    
def hover(event):    
    visible = annotation.get_visible()
    if event.inaxes == plt.gca():
        contains, ind = scatterPlot.contains(event)
        if contains:
            update_annotation(ind)
            annotation.set_visible(True)
            plt.gcf().canvas.draw_idle()
        else:
            if visible:
                annotation.set_visible(False)
                plt.gcf().canvas.draw_idle()

plt.gcf().canvas.mpl_connect("motion_notify_event", hover)

plt.show()
```
