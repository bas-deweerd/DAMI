```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(1)
%matplotlib notebook

x = np.random.rand(15)
y = np.random.rand(15)
names = np.array(list("ABCDEFGHIJKLMNO"))
c = np.random.randint(1,5,size=15)

norm = plt.Normalize(1,4)
colormap = plt.cm.RdYlGn

figure,axes = plt.subplots()
scatterPlot = plt.scatter(x,y,c=c, s=100, cmap=colormap, norm=norm)

annotation = axes.annotate("", xy=(0,0), xytext=(20,20),textcoords="offset points",
                    bbox=dict(boxstyle="round", fc="w"),
                    arrowprops=dict(arrowstyle="->"))
annotation.set_visible(False)

def update_annotation(ind):

    position = scatterPlot.get_offsets()[ind["ind"][0]]
    annotation.xy = position
    text = "{}, {}".format(" ".join(list(map(str,ind["ind"]))), 
                           " ".join([names[n] for n in ind["ind"]]))
    annotation.set_text(text)
    annotation.get_bbox_patch().set_facecolor(cmap(norm(c[ind["ind"][0]])))
    annotation.get_bbox_patch().set_alpha(0.4)


def hover(event):
    visible = annotation.get_visible()
    if event.inaxes == axes:
        continu, ind = scatterPlot.contains(event)
        if continu:
            update_annotation(ind)
            annotation.set_visible(True)
            figure.canvas.draw_idle()
        else:
            if visible:
                annotation.set_visible(False)
                figure.canvas.draw_idle()

figure.canvas.mpl_connect("motion_notify_event", hover)

plt.show()
```
